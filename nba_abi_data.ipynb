```python
# Import necessary libraries
from nba_api.live.nba.endpoints import scoreboard, boxscore
import pandas as pd

# Step 1: Fetch today's games
games = scoreboard.ScoreBoard()
games_dict = games.get_dict()
game_ids = [game['gameId'] for game in games_dict['scoreboard']['games']]

# Step 2: Initialize an empty list to hold DataFrames
all_player_stats = []

# Step 3: Loop through each game and fetch player stats
for game_id in game_ids:
    # Fetch box score for the game
    game_stats = boxscore.BoxScore(game_id)
    game_stats_dict = game_stats.get_dict()

    # Extract home and away player stats
    home_players = game_stats_dict['game']['homeTeam']['players']
    away_players = game_stats_dict['game']['awayTeam']['players']

    # Convert to DataFrame and normalize the 'statistics' column
    home_df = pd.DataFrame(home_players)
    away_df = pd.DataFrame(away_players)

    # Expand the 'statistics' column into separate columns
    if 'statistics' in home_df.columns:
        home_stats = pd.json_normalize(home_df['statistics'])
        home_df = pd.concat([home_df.drop(columns=['statistics']), home_stats], axis=1)

    if 'statistics' in away_df.columns:
        away_stats = pd.json_normalize(away_df['statistics'])
        away_df = pd.concat([away_df.drop(columns=['statistics']), away_stats], axis=1)

    # Add columns for team and game ID
    home_df['team'] = game_stats_dict['game']['homeTeam']['teamName']
    home_df['game_id'] = game_id

    away_df['team'] = game_stats_dict['game']['awayTeam']['teamName']
    away_df['game_id'] = game_id

    # Append to the list
    all_player_stats.append(home_df)
    all_player_stats.append(away_df)

# Step 4: Concatenate all DataFrames into a single DataFrame
if all_player_stats:
    combined_stats_df = pd.concat(all_player_stats, ignore_index=True)
else:
    combined_stats_df = pd.DataFrame()  # No games today

# combined_stats_df now contains all player stats from today's games with separate columns for each stat
```

```python
def add_doubles(df):
    dd = [0 for i in range(df.shape[0])]
    td = [0 for i in range(df.shape[0])]
    for i in (range(df.shape[0])):
        doubles_count = 0
        check_doubles = ['points', 'reboundsTotal', 'assists', 'steals', 'blocks']
        for stat in check_doubles:
            if df.loc[i, stat] >= 10:
                doubles_count += 1
        if doubles_count >= 2:
            dd[i] = 1
        if doubles_count >= 3:
            td[i] = 1
    df['DD'] = dd
    df['TD'] = td


def calculate_FPTS(df):
    multipliers = {
        'points': 1,
        'threePointersMade': 0.5,
        'reboundsTotal': 1.25,
        'assists': 1.5,
        'steals': 2,
        'blocks': 2,
        'turnovers': -0.5
    }
    indices = len(df)
    fpts_list = []
    for i in (range(indices)):
        fpts = 0
        doubles = 0
        for stat, multiplier in multipliers.items():
            if stat in ['points', 'reboundsTotal', 'assists', 'steals', 'blocks']:
                if df.loc[i, stat] >= 10:
                    doubles += 1
            fpts += df.loc[i, stat] * multiplier
        if doubles >= 2:
            fpts += 1.5
        if doubles >= 3:
            fpts += 3
        fpts_list.append(fpts)
    return fpts_list
```

```python
add_doubles(combined_stats_df)
combined_stats_df['FPTS'] = calculate_FPTS(combined_stats_df)
```
